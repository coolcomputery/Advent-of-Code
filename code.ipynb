```python
#5a*
fin=open('in.txt','r')
L=fin.read().split('\n')
lines=[]
for l in L:
    ends=l.split(' -> ')
    lines.append(list(map(lambda e:list(map(int,e.split(','))),ends)))
freq={}
def add(p):
    freq[p]=1+(freq[p] if p in freq else 0)
for l in lines:
    x0,y0,x1,y1=l[0][0],l[0][1],l[1][0],l[1][1]
    #print(x0,y0,x1,y1)
    if x0==x1:
        for y in range(min(y0,y1),max(y0,y1)+1):
            add((x0,y))
    elif y0==y1:
        for x in range(min(x0,x1),max(x0,x1)+1):
            add((x,y0))
cnt=0
for p in freq:
    if freq[p]>1:
        cnt+=1
print(cnt)
```

```python
#5b
fin=open('in.txt','r')
L=fin.read().split('\n')
lines=[]
for l in L:
    ends=l.split(' -> ')
    lines.append(list(map(lambda e:list(map(int,e.split(','))),ends)))
freq={}
def add(p):
    freq[p]=1+(freq[p] if p in freq else 0)
for l in lines:
    x0,y0,x1,y1=l[0][0],l[0][1],l[1][0],l[1][1]
    #print(x0,y0,x1,y1)
    if x0==x1:
        for y in range(min(y0,y1),max(y0,y1)+1):
            add((x0,y))
    elif y0==y1:
        for x in range(min(x0,x1),max(x0,x1)+1):
            add((x,y0))
    else:
        dx,dy=1 if x0<x1 else -1,1 if y0<y1 else -1
        for s in range(max(y0,y1)+1-min(y0,y1)):
            add((x0+dx*s,y0+dy*s))
cnt=0
for p in freq:
    if freq[p]>1:
        cnt+=1
print(cnt)
```

```python
#8b
import itertools
fin=open('in.txt','r')
masks='1110111,0010010,1011101,1011011,0111010,1101011,1101111,1010010,1111111,1111011'.split(',')
masks={masks[i]:i for i in range(10)}
print(masks)
L=fin.read().split('\n')
cnt=0
def tomask(s):
    m=[0]*7
    for c in s:
        m[ord(c)-ord('a')]+=1
    return m
for l in L:
    P=tuple(map(lambda s:s.split(' '),l.split(' | ')))
    Ms=[]
    for s in P[0]:
        Ms.append(tomask(s))
    #print(P)
    #print(Ms)
    ret=None
    for perm in itertools.permutations(range(7)):
        good=True
        V={}
        for m in Ms:
            nm=''.join([str(m[perm[i]]) for i in range(7)])
            if not(nm in masks):
                good=False
                break
            else:
                V[nm]=masks[nm]
        if good:
            v=''
            for s in P[1]:
                m=tomask(s)
                nm=''.join([str(m[perm[i]]) for i in range(7)])
                v+=str(V[nm])
            cnt+=int(v)
            #print(v)
            break
print(cnt)
```

```python
#9b
fin=open('in.txt','r')
A=list(map(lambda s:list(map(int,list(s))),fin.read().split('\n')))
R,C=len(A),len(A[0])
dr,dc=[1,0,-1,0],[0,1,0,-1]
s=0
dest=[[None]*C for r in range(R)]
for r in range(len(A)):
    for c in range(len(A[r])):
        if A[r][c]!=9:
            bpt,bscr=None,A[r][c]
            for d in range(4):
                nr,nc=r+dr[d],c+dc[d]
                if 0<=nr and nr<len(A) and 0<=nc and nc<len(A[r]) and A[nr][nc]<bscr:
                    bpt,bscr=(nr,nc),A[nr][nc]
            dest[r][c]=bpt
par=[i for i in range(R*C)]
szs=[1]*(R*C)
def find(i):
    if i==par[i]:
        return i
    else:
        par[i]=find(par[i])
        return par[i]
def merge(i,j):
    fi,fj=find(i),find(j)
    par[fi]=fj
    szs[fj]+=szs[fi]
    szs[fi]=0
for r in range(R):
    for c in range(C):
        if A[r][c]!=9:
            i=r*C+c
            p=dest[r][c]
            if p!=None:
                merge(i,p[0]*C+p[1])
S=sorted(szs)
p=1
for v in S[-3:]:
    p*=v
p
```
