```python
#18a,b
lines=open('in.txt','r').read().split('\n')
class Node:
    def __init__(self,v,p,cs):
        self.val=v
        self.parent=p
        self.children=cs
    def add_child(self,c):
        self.children.append(c)
        c.parent=self
    def __str__(self):
        return '['+(','.join(map(str,self.children)))+']' if self.val==None else str(self.val)
    def isnum(self):
        return self.val!=None
def tree(s):
    stack=[]
    root=None
    i=0
    while i<len(s):
        #print(list(map(str,stack)))
        if s[i]=='[':
            n=Node(None,None,[])
            if root==None:
                root=n
            if len(stack)>0:
                p=stack[-1]
                p.add_child(n)
            stack.append(n)
            i+=1
        elif s[i]==']':
            stack.pop()
            i+=1
        elif s[i]==',':
            i+=1
        else:
            j=i
            while '0'<=s[j] and s[j]<='9':
                j+=1
            n=Node(int(s[i:j]),None,[])
            p=stack[-1]
            p.add_child(n)
            i=j
    return root
def printtree(n,d=0):
    print(''*d+str(n.val))
    for c in n.children:
        printtree(c,d+1)
def add(T,U):
    n=Node(None,None,[])
    n.add_child(T)
    n.add_child(U)
    return n
def explode(T):
    STOP=[False]
    def dfs(n,d):
        #print(str(n))
        if STOP[0]:
            return
        if not(n.isnum()):
            l,r=n.children
            if (l.isnum()
                and r.isnum()
                and d>=4):
                #can a pair of pairs explode? (apparently no)
                a,b=l.val,r.val
                #add to left
                anc,pd=n,d
                while pd>0:
                    p=anc.parent
                    if p.children[0]!=anc:
                        nn=p.children[0]
                        while not(nn.isnum()):
                            nn=nn.children[1]
                        nn.val+=a
                        break
                    anc,pd=p,pd-1
                #add to right
                anc,pd=n,d
                while pd>0:
                    p=anc.parent
                    if p.children[1]!=anc:
                        nn=p.children[1]
                        while not(nn.isnum()):
                            nn=nn.children[0]
                        nn.val+=b
                        break
                    anc,pd=p,pd-1
                #replace node
                p=n.parent
                p.children[p.children.index(n)]=Node(0,p,[])
                STOP[0]=True
                return
        for c in n.children:
            dfs(c,d+1)
    dfs(T,0)
    return STOP[0]
def split(T):
    STOP=[False]
    def dfs(n,d):
        #print(str(n))
        if STOP[0]:
            return
        if n.isnum() and n.val>=10:
            v=n.val
            l,r=Node(v//2,None,[]),Node((v+1)//2,None,[])
            n.add_child(l)
            n.add_child(r)
            n.val=None
            STOP[0]=True
            return
        for c in n.children:
            dfs(c,d+1)
    dfs(T,0)
    return STOP[0]
def reduce(T):
    while explode(T) or split(T):
        pass
def mag(T):
    return (T.val if T.isnum()
            else 3*mag(T.children[0])+2*mag(T.children[1]))
T=tree(lines[0])
for l in lines[1:]:
    T=add(T,tree(l))
    reduce(T)
print(mag(T))
m=0
for a in lines:
    for b in lines:
        T=add(tree(a),tree(b))
        reduce(T)
        m=max(m,mag(T))
print(m)
```
