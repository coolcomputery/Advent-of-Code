```python
#14b
def add(F,k,v):
    F[k]=(F[k] if k in F else 0)+v
lines=open('in.txt','r').read().split('\n')
S0=lines[0]
L,R=S0[0],S0[-1]
SF={}
for i in range(len(S0)-1):
    p=S0[i:i+2]
    add(SF,p,1)
print(SF)
rules={}
for l in lines[2:]:
    a,b=l.split(' -> ')
    rules[a]=b
for rep in range(40):
    nSF=dict(SF)
    for p in SF:
        f=SF[p]
        if p in rules:
            v=rules[p]
            l,r=p[0]+v,v+p[1]
            add(nSF,l,f)
            add(nSF,r,f)
            add(nSF,p,-f)
    SF=nSF
F={}
for p in SF:
    add(F,p[0],SF[p])
    add(F,p[1],SF[p])
add(F,L,1)
add(F,R,1)
for p in F:
    F[p]//=2
P=sorted([(F[a],a) for a in F])
print(P)
P[-1][0]-P[0][0]
```
